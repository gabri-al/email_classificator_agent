## Persist Data
Store above examples as UC Delta tables and sync delta tables to Lakebase ([doc](https://docs.databricks.com/aws/en/oltp/instances/sync-data/sync-table?language=Python+SDK))

(Executing this section only when the data needs to be recreated).

```python
## Generate delta tables

if regenerate_data:
  
  # Set up catalog, schemas
  spark.sql("CREATE CATALOG IF NOT EXISTS "+catalog_)
  spark.sql("USE CATALOG "+catalog_)
  spark.sql("CREATE SCHEMA IF NOT EXISTS "+schema_)
  spark.sql("USE SCHEMA "+schema_)

  # Drop & Create tables
  spark.sql("DROP TABLE IF EXISTS classificator_agent_customers")
  spark.sql("DROP TABLE IF EXISTS classificator_agent_orders")
  spark.sql("DROP TABLE IF EXISTS classificator_agent_tickets")
  spark.sql("DROP TABLE IF EXISTS classificator_agent_emails")

  df_customer_example.write.mode("overwrite").saveAsTable("classificator_agent_customers")
  df_order_example.write.mode("overwrite").saveAsTable("classificator_agent_orders")
  df_tickets_example.write.mode("overwrite").saveAsTable("classificator_agent_tickets")
  df_emails_example.write.mode("overwrite").saveAsTable("classificator_agent_emails")

  # Add Primary Keys
  spark.sql(
  """
    ALTER TABLE classificator_agent_customers ALTER COLUMN Customer_Id SET NOT NULL
  """)
  spark.sql(
  """
    ALTER TABLE classificator_agent_customers ADD CONSTRAINT customers_pk PRIMARY KEY (Customer_Id)
  """)
  
  spark.sql(
  """
    ALTER TABLE classificator_agent_orders ALTER COLUMN Order_Id SET NOT NULL
  """)
  spark.sql(
  """
    ALTER TABLE classificator_agent_orders ADD CONSTRAINT orders_pk PRIMARY KEY (Order_Id)
  """)

  spark.sql(
  """
    ALTER TABLE classificator_agent_tickets ALTER COLUMN Ticket_Nr SET NOT NULL
  """)
  spark.sql(
  """
    ALTER TABLE classificator_agent_tickets ADD CONSTRAINT tickets_pk PRIMARY KEY (Ticket_Nr)
  """)

  spark.sql(
  """
    ALTER TABLE classificator_agent_emails ALTER COLUMN Email_Id SET NOT NULL
  """)
  spark.sql(
  """
    ALTER TABLE classificator_agent_emails ADD CONSTRAINT emails_pk PRIMARY KEY (Email_Id)
  """)

else:
  spark.sql("USE CATALOG "+catalog_)
  spark.sql("USE SCHEMA "+schema_)
```

```python
## Sync Delta tables with Lakebase Instance

from databricks.sdk import WorkspaceClient
from databricks.sdk.service.database import DatabaseInstance
from databricks.sdk.service.database import SyncedDatabaseTable, SyncedTableSpec, NewPipelineSpec, SyncedTableSchedulingPolicy

## Helper function
def clean_lakebase(delta_table_name:str, lakebase_instance:str):
    w = WorkspaceClient()
    synced_table_name = delta_table_name+"_synced"

    # Drop synced table
    try:
        #spark.sql("DROP TABLE IF EXISTS "+catalog_+"."+schema_+"."+synced_table_name)
        w.database.delete_synced_database_table(name=catalog_+"."+schema_+"."+synced_table_name)
    except:
        print("Nothing to drop on UC")
    # Drop Lakebase table
    try:
        spark.sql("DROP TABLE IF EXISTS `" + "databricks_postgres." + schema_ + "." + synced_table_name + "`")
        # Or connect to Lakebase and clean tables manually
    except:
        print("Nothing to drop on Lakebase")


def sync_delta_to_lakebase(delta_table_name:str, primary_keys:[str], lakebase_instance:str):
  w = WorkspaceClient()
  synced_table_name = delta_table_name+"_synced"
  
  # Create a synced table in a standard UC catalog
  synced_table = w.database.create_synced_database_table(
      SyncedDatabaseTable(
            name=catalog_+"."+schema_+"."+synced_table_name,  # Full three-part name
            database_instance_name=lakebase_instance,  # Required for standard catalogs
            logical_database_name="databricks_postgres",  # Required for standard catalogs
            spec=SyncedTableSpec(
                source_table_full_name=catalog_+"."+schema_+"."+delta_table_name,
                primary_key_columns=primary_keys,
                scheduling_policy=SyncedTableSchedulingPolicy.SNAPSHOT,
                create_database_objects_if_missing=True,  # Create database/schema if needed
                new_pipeline_spec=NewPipelineSpec(
                    storage_catalog=catalog_,
                    storage_schema=schema_
                )
            ),
      )
  )
  print(f"Created synced table: {synced_table.name}")

  status = w.database.get_synced_database_table(name=catalog_+"."+schema_+"."+synced_table_name)
  print(f"Synced table status: {status.data_synchronization_status.detailed_state}")
  print(f"Status message: {status.data_synchronization_status.message}")
```

```python
## Run function on tables
if regenerate_lakebase_data:
  clean_lakebase("classificator_agent_customers", lakebase_instance_)
  clean_lakebase("classificator_agent_orders", lakebase_instance_)
  clean_lakebase("classificator_agent_tickets", lakebase_instance_)

if regenerate_lakebase_data:
  sync_delta_to_lakebase("classificator_agent_customers", ["Customer_Id"], lakebase_instance_)
  sync_delta_to_lakebase("classificator_agent_orders", ["Order_Id"], lakebase_instance_)
  sync_delta_to_lakebase("classificator_agent_tickets", ["Ticket_Nr"], lakebase_instance_)
```

## Create the Retriever functions

Create UC functions that can be used as retriever tools by the agent ([doc](https://docs.databricks.com/aws/en/generative-ai/agent-framework/structured-retrieval-tools), Databricks Demos [example in Notebook 5.1](https://notebooks.databricks.com/demos/lakehouse-iot-platform/index.html#))

```python
spark.sql("USE CATALOG "+catalog_)
spark.sql("USE SCHEMA "+schema_)
```

```python
%sql
DROP FUNCTION IF EXISTS classificator_agent_customer_retriever;
DROP FUNCTION IF EXISTS classificator_agent_order_retriever;
DROP FUNCTION IF EXISTS classificator_agent_ticket_retriever;

CREATE OR REPLACE FUNCTION classificator_agent_customer_retriever(customer_email STRING)
RETURNS STRUCT<Customer_Id BIGINT, Name STRING, Surname STRING, Email STRING, Address STRING>
LANGUAGE SQL
COMMENT 'Given a customer email address, look up this customer in the customers table and return a single struct with fields: Customer_Id, Name, Surname, Email, Address. If no customer is found, return NULL.'
RETURN (
  SELECT struct(Customer_Id, Name, Surname, Email, Address)
  FROM classificator_agent_customers_synced
  WHERE Email = classificator_agent_customer_retriever.customer_email
  LIMIT 1
);

CREATE OR REPLACE FUNCTION classificator_agent_order_retriever(customer_id BIGINT)
RETURNS STRUCT<Order_Id STRING, Created STRING, Cart STRING, Amount DOUBLE, Customer_Id BIGINT>
LANGUAGE SQL
COMMENT 'Given a customer ID, look up his/her records in the orders table and return a single struct with fields: Order_Id, Created, Cart, Amount, Customer_Id for that order ID. If no order is found, return NULL.'
RETURN (
  SELECT struct(Order_Id, Created, Cart, Amount, Customer_Id)
  FROM classificator_agent_orders_synced
  WHERE Customer_Id = classificator_agent_order_retriever.customer_id
  ORDER BY Created DESC
  LIMIT 1
);

CREATE OR REPLACE FUNCTION classificator_agent_ticket_retriever(customer_id BIGINT)
RETURNS STRUCT<Customer_Id BIGINT, Ticket_Nr STRING, Created STRING, Description STRING, Label STRING>
LANGUAGE SQL
COMMENT 'Given a customer ID, look up this customer in the tickets table and return a single struct with fields: Customer_Id, Ticket_Nr, Created, Description, Label for the most recent ticket of that customer. If no ticket is found, return NULL.'
RETURN (
  SELECT struct(Customer_Id, Ticket_Nr, Created, Description, Label)
  FROM classificator_agent_tickets_synced
  WHERE Customer_Id = classificator_agent_ticket_retriever.customer_id
  ORDER BY Created DESC
  LIMIT 1
);
```
